3.Напишите программу, которая по координатам точки выводит сообщение в какой четверти лежит эта точка (1-я, 2-я, 3-я или 4-я четверти).

Принадлежность точки с координатами (x,y) к одной из четвертей определяется следующим образом:

* для первой четверти x ≥ 0 и y ≥ 0;

* для второй четверти x ≤ 0 и y ≥ 0;

* для третьей четверти x ≤ 0 и y ≤ 0;

* для четвертой четверти x ≥ 0 и y ≤ 0.

Проверьте делится ли без остатка предпоследнее число последовательности. Если делится, то выведите 1, если нет то 0.
